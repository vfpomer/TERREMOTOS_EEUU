```python
#calcular media
# Lista de columnas a analizar
columnas = ['nst', 'gap', 'dmin', 'horizontalError', 'magError', 'magNst']

# Iterar sobre cada columna y calcular la media
for col in columnas:
    datos = df[col].dropna()  # Ignora los valores nulos
    if datos.empty:
        print(f"Media de '{col}': no se puede calcular (todos los valores son nulos)")
    else:
        media = datos.mean()
        print(f"Media de '{col}': {media:.2f}")
```

```python
#calcular moda
# Lista de columnas
columnas = ['nst', 'gap', 'dmin', 'horizontalError', 'magError', 'magNst']

for col in columnas:
    datos = pd.Series(df[col])
    moda = datos.mode()
    if not moda.empty:
        print(f"Moda {col}:", moda[0])
    else:
        print(f"Moda {col}: No hay moda")
```

```python
#calcular mediana
# Lista de columnas
columnas = ['nst', 'gap', 'dmin', 'horizontalError', 'magError', 'magNst']

for col in columnas:
    datos = pd.Series(df[col])
    mediana = datos.median()
    print(f"Mediana {col}:", mediana)
```

```python
#como arreglo los valores nulos sin cargarnos los datos originales
df_limpio = df.copy() #copia del dataframe original
```

```python
df_limpio.head(3) #verifico que la copia se haya hecho bien
```

```python
# Como hay entre un 14,5% y un 20% de datos nulos, creo que la mejor manera de rellenar es 
# rellenando con 'Sin datos'.
# #Sustitución en  de valores faltantes por 'Sin datos'
df_limpio['nst'] = df_limpio['nst'].fillna('Sin datos') #relleno los valores nulos de nst con 'Sin datos'
df_limpio['gap'] = df_limpio['gap'].fillna('Sin datos') #relleno los valores nulos de gap con 'Sin datos'
df_limpio['dmin'] = df_limpio['dmin'].fillna('Sin datos') #relleno los valores nulos de dmin con 'Sin datos'
df_limpio['horizontalError'] = df_limpio['horizontalError'].fillna('Sin datos') #relleno los valores nulos de horizontalError con 'Sin datos'
df_limpio['magError'] = df_limpio['magError'].fillna('Sin datos') #relleno los valores nulos de magError con 'Sin datos'
df_limpio['magNst'] = df_limpio['magNst'].fillna('Sin datos') #relleno los valores nulos de magNst con 'Sin datos'
```
